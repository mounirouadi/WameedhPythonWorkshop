# Tuples

```python
#Tuples
Tuple_1 = ("Phone", "Wallet", "Tshirt")
Tuple_2 = (1, 5, 7, 9, 3)
Tuple_3 = (True, False, False)
print(Tuple_1,Tuple_2,Tuple_3)
print(type(Tuple_1))

#Mixed Tuples
Tuple = ("abc", 34, True, 40, "male")
print(len(Tuple))
```

```python
#Access Tuples
fruits=("bannana","orange","strawberry","apple","dates")

print(fruits[1])
print(fruits[4])
print(fruits[1:4])
print(fruits[:5])
print(fruits[2:])
```

```python
# Changing Tuple item
fruits=("bannana","orange","strawberry","apple","dates")
fruits[3]="Mounir"
print(fruits)
```

```python
# Changing Tuple item 
# Convert the tuple into a list to be able to change it
fruits_tuple=("bannana","orange","strawberry","apple","dates")
fruits_list=list(fruits_tuple)
fruits_list[2]="kiwi"
fruits_tuple=tuple(fruits_list)
print(fruits_tuple)
```

```python
# Adding items to tuples
fruits=("bannana","orange","strawberry","apple","dates")
fruits_list=list(fruits)
fruits_list.append("kiwi")
fruits=tuple(fruits_list)
print(fruits)

# Add a tuple to a tuple
fruits=("bannana","orange","strawberry","apple","dates")
item = ("pear",)
fruits+= item
print(fruits)

# Removing an item from tuples
fruits=("bannana","orange","strawberry","apple","dates")
fruits_list=list(fruits)
fruits_list.remove("dates")
fruits=tuple(fruits_list)
print(fruits)

# Adding tuples
fruits=("bannana","orange","strawberry","apple","dates")
phones=("Iphone","Samsung","Huawei")
print(fruits+phones)
```

```python
#Summing it in a function
def csv_parser(file_directory):
    #Opening our CSV file
    f = open(file_directory, 'r')
    #Reading the data and splitting on new lines \n
    data = f.read().splitlines()
    
    splitted_data = []
    for i in data:
        #Splitting each line on where the Comma (,) and making it a list
        x = i.split(',')
        for j, element in enumerate(x):
            #Replacing the empty values with '0'
            if(element == ''):
                x[j] = '0'
        #Append the splitted and cleaned lines on a new list
        splitted_data.append(x)
    #Initializing a new dictionary
    data_dictionary = {}
    #Adding empty lists with keys of our CSV headers
    for i in range(len(splitted_data[0])):
        data_dictionary[splitted_data[0][i]] = []
    #Appending each Row to its corresponding key
    for line in splitted_data[1:]:
        for i in range(len(splitted_data[0])):
            data_dictionary[splitted_data[0][i]].append(float(line[i]))
    #Returning our data dictionary to the main code
    return data_dictionary
```

```python
csv_parser("data.csv")
```
